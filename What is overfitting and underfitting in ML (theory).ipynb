# What is underfitting in machine learning

```python
Here are five key points about underfitting in machine learning:

1. **Definition**: Underfitting occurs when a model is too simple to capture the underlying patterns in the data,
                   leading to poor performance on both training and test datasets.

2. **High Bias**: An underfit model has high bias, meaning it makes overly simplistic assumptions about the data,
                  resulting in inaccurate predictions.

3. **Poor Generalization**: The model fails to generalize well to new, unseen data, 
                            as it hasn't learned the complexities of the training data.

4. **Causes**: Common causes include using a model that is too simple, insufficient training data,
               or excessive regularization.

5. **Solutions**: To address underfitting, you can increase model complexity, add more features, 
                  reduce regularization, or gather more training data.
```

# What is overfitting in machine learning

```python
 Here are five key points about overfitting in machine learning:

1. **Definition**: Overfitting occurs when a model learns the training data too well, including noise and outliers,
                   resulting in poor performance on new, unseen data.

2. **High Variance**: An overfit model has high variance, meaning it performs exceptionally well on training data 
                      but poorly on test data.

3. **Poor Generalization**: The model fails to generalize to new data, making it unreliable for real-world applications.

4. **Causes**: Common causes include overly complex models, too many features, and insufficient regularization.

5. **Solutions**: To mitigate overfitting, you can simplify the model, use cross-validation, 
                  apply regularization techniques, or gather more training data.
```
